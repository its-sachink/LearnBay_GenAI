```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load dataset
tips = sns.load_dataset("tips")

# 1. Boxplot & Histogram
sns.boxplot(x=tips["total_bill"])
plt.show()
sns.histplot(tips["total_bill"], kde=True)
plt.show()

# 2. Skewness & Kurtosis
print("Skewness:", stats.skew(tips["total_bill"]))
print("Kurtosis:", stats.kurtosis(tips["total_bill"]))

# 3. IQR Method
Q1 = tips["total_bill"].quantile(0.25)
Q3 = tips["total_bill"].quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = tips[(tips["total_bill"] < Q1 - 1.5*IQR) | (tips["total_bill"] > Q3 + 1.5*IQR)]

# 4. Z-Score Method
z_scores = stats.zscore(tips["total_bill"])
outliers_z = tips[(abs(z_scores) > 3)]

# 5. Correlation
print("Correlation (total_bill vs size):", tips["total_bill"].corr(tips["size"]))
```
